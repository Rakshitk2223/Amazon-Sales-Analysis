```python
# Import additional libraries for comprehensive analysis
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from datetime import datetime
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Set up plotting style
plt.style.use('default')
sns.set_palette("husl")
warnings.filterwarnings('ignore')

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
```

```python
# 1. BASIC DATA INFORMATION
print("="*60)
print("AMAZON SALES DATASET - INITIAL ANALYSIS")
print("="*60)
print(f"Dataset Shape: {df.shape}")
print(f"Memory Usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
print("\nColumn Information:")
print("-"*40)
df.info()
```

```python
# 2. MISSING VALUES AND DUPLICATES ANALYSIS
print("\n" + "="*60)
print("MISSING VALUES ANALYSIS")
print("="*60)
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100

missing_df = pd.DataFrame({
    'Column': missing_values.index,
    'Missing Count': missing_values.values,
    'Missing Percentage': missing_percentage.values
}).sort_values('Missing Count', ascending=False)

print(missing_df[missing_df['Missing Count'] > 0])

print(f"\nDuplicate Rows: {df.duplicated().sum()}")
print(f"Unique Products: {df['product_id'].nunique()}")
print(f"Unique Users: {df['user_id'].nunique()}")
print(f"Unique Reviews: {df['review_id'].nunique()}")
```

```python
# 3. DATA EXAMINATION AND TYPE CONVERSION
print("\n" + "="*60)
print("DATA EXAMINATION")
print("="*60)

# First examine the data formats
print("Sample data from key columns:")
print("Discounted Price samples:", df['discounted_price'].head())
print("Actual Price samples:", df['actual_price'].head())
print("Discount Percentage samples:", df['discount_percentage'].head())
print("Rating samples:", df['rating'].head())
print("Rating Count samples:", df['rating_count'].head())

# Check for any problematic values
print("\nUnique values in rating column:", df['rating'].unique()[:10])
print("Unique values in discount_percentage:", df['discount_percentage'].unique()[:10])
```

```python
# 4. CATEGORICAL DATA ANALYSIS
print("\n" + "="*60)
print("CATEGORICAL DATA ANALYSIS")
print("="*60)

categorical_cols = df.select_dtypes(include=['object']).columns
print("Categorical Columns:", list(categorical_cols))

print(f"\nTop 10 Categories:")
print(df['category'].value_counts().head(10))

print(f"\nTop 10 Most Reviewed Products:")
top_products = df.groupby('product_name')['rating_count'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
# 5. COMPREHENSIVE DATA ANALYSIS WITH PROPER CONVERSION
print("\n" + "="*60)
print("COMPREHENSIVE DATA ANALYSIS")
print("="*60)

# Clean and convert price columns
df['discounted_price_clean'] = pd.to_numeric(df['discounted_price'].str.replace('₹', '').str.replace(',', ''), errors='coerce')
df['actual_price_clean'] = pd.to_numeric(df['actual_price'].str.replace('₹', '').str.replace(',', ''), errors='coerce')
df['discount_percentage_clean'] = pd.to_numeric(df['discount_percentage'].str.replace('%', ''), errors='coerce')

# Clean rating data - handle the "|" issue
df['rating_clean'] = pd.to_numeric(df['rating'].astype(str).str.split('|').str[0], errors='coerce')

# Clean rating count - remove commas and convert to numeric
df['rating_count_clean'] = pd.to_numeric(df['rating_count'].astype(str).str.replace(',', ''), errors='coerce')

print("PRICE ANALYSIS:")
print("-" * 30)
print(f"Average Discounted Price: ₹{df['discounted_price_clean'].mean():.2f}")
print(f"Median Discounted Price: ₹{df['discounted_price_clean'].median():.2f}")
print(f"Average Actual Price: ₹{df['actual_price_clean'].mean():.2f}")
print(f"Median Actual Price: ₹{df['actual_price_clean'].median():.2f}")
print(f"Average Discount: {df['discount_percentage_clean'].mean():.2f}%")
print(f"Max Discount: {df['discount_percentage_clean'].max():.0f}%")

print(f"\nRATING ANALYSIS:")
print("-" * 30)
print(f"Average Rating: {df['rating_clean'].mean():.2f}")
print(f"Median Rating: {df['rating_clean'].median():.2f}")
print(f"Rating Range: {df['rating_clean'].min():.1f} - {df['rating_clean'].max():.1f}")
print(f"Average Rating Count: {df['rating_count_clean'].mean():.0f}")
print(f"Median Rating Count: {df['rating_count_clean'].median():.0f}")

print(f"\nTOP 5 CATEGORIES BY PRODUCT COUNT:")
print("-" * 40)
top_categories = df['category'].value_counts().head(5)
for idx, (category, count) in enumerate(top_categories.items(), 1):
    print(f"{idx}. {category.split('|')[-1]}: {count} products")

print(f"\nPRICE RANGES:")
print("-" * 20)
price_ranges = pd.cut(df['discounted_price_clean'], bins=5, precision=0)
print(price_ranges.value_counts().sort_index())
```

```python
# 6. INITIAL VISUALIZATIONS
print("\n" + "="*60)
print("CREATING INITIAL VISUALIZATIONS")
print("="*60)

# Create a figure with multiple subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Amazon Sales Dataset - Initial Analysis', fontsize=16, fontweight='bold')

# 1. Price Distribution
axes[0, 0].hist(df['discounted_price_clean'], bins=30, color='skyblue', alpha=0.7, edgecolor='black')
axes[0, 0].set_title('Distribution of Discounted Prices')
axes[0, 0].set_xlabel('Price (₹)')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].ticklabel_format(style='plain', axis='x')

# 2. Rating Distribution
rating_counts = df['rating_clean'].value_counts().sort_index()
axes[0, 1].bar(rating_counts.index, rating_counts.values, color='lightgreen', alpha=0.7, edgecolor='black')
axes[0, 1].set_title('Distribution of Ratings')
axes[0, 1].set_xlabel('Rating')
axes[0, 1].set_ylabel('Count')

# 3. Discount Percentage Distribution
axes[1, 0].hist(df['discount_percentage_clean'], bins=20, color='coral', alpha=0.7, edgecolor='black')
axes[1, 0].set_title('Distribution of Discount Percentages')
axes[1, 0].set_xlabel('Discount (%)')
axes[1, 0].set_ylabel('Frequency')

# 4. Top 10 Categories
top_cats = df['category'].str.split('|').str[-1].value_counts().head(10)
axes[1, 1].barh(range(len(top_cats)), top_cats.values, color='gold', alpha=0.7, edgecolor='black')
axes[1, 1].set_yticks(range(len(top_cats)))
axes[1, 1].set_yticklabels(top_cats.index, fontsize=8)
axes[1, 1].set_title('Top 10 Product Categories')
axes[1, 1].set_xlabel('Count')

plt.tight_layout()
plt.show()

# Summary statistics table
print("\nSUMMARY STATISTICS TABLE:")
print("=" * 50)
summary_stats = pd.DataFrame({
    'Metric': ['Total Products', 'Total Reviews', 'Avg Rating', 'Avg Discount %', 
               'Avg Discounted Price', 'Most Popular Category', 'Price Range'],
    'Value': [
        f"{df.shape[0]:,}",
        f"{df['rating_count_clean'].sum():,.0f}",
        f"{df['rating_clean'].mean():.2f}",
        f"{df['discount_percentage_clean'].mean():.1f}%",
        f"₹{df['discounted_price_clean'].mean():,.2f}",
        df['category'].str.split('|').str[-1].value_counts().index[0],
        f"₹{df['discounted_price_clean'].min():.0f} - ₹{df['discounted_price_clean'].max():,.0f}"
    ]
})
print(summary_stats.to_string(index=False))
```
