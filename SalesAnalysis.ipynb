```python
# 7. MISSING VALUE HANDLING
print("\n" + "="*60)
print("MISSING VALUE HANDLING")
print("="*60)

# Create a copy of the dataframe for cleaning
df_clean = df.copy()

# Handle missing values based on data type
print("Handling missing values...")

# For numerical columns - use mean
numerical_cols = ['discounted_price_clean', 'actual_price_clean', 'discount_percentage_clean', 'rating_clean', 'rating_count_clean']
for col in numerical_cols:
    if df_clean[col].isnull().sum() > 0:
        mean_value = df_clean[col].mean()
        df_clean[col].fillna(mean_value, inplace=True)
        print(f"Filled {col} missing values with mean: {mean_value:.2f}")

# For categorical columns - use mode
categorical_cols = ['category', 'product_name', 'about_product', 'user_name', 'review_title', 'review_content']
for col in categorical_cols:
    if df_clean[col].isnull().sum() > 0:
        mode_value = df_clean[col].mode()[0] if not df_clean[col].mode().empty else 'Unknown'
        df_clean[col].fillna(mode_value, inplace=True)
        print(f"Filled {col} missing values with mode: {mode_value}")

# Verify no missing values remain
print(f"\nMissing values after cleaning:")
missing_after = df_clean.isnull().sum()
print(missing_after[missing_after > 0])
if missing_after.sum() == 0:
    print("✅ All missing values have been handled successfully!")
```

```python
# 8. OUTLIER DETECTION AND ANALYSIS
print("\n" + "="*60)
print("OUTLIER DETECTION AND ANALYSIS")
print("="*60)

# Function to detect outliers using IQR method
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Detect outliers in key numerical columns
outlier_columns = ['discounted_price_clean', 'actual_price_clean', 'rating_count_clean']

print("Outlier Analysis:")
print("-" * 30)

for col in outlier_columns:
    outliers, lower, upper = detect_outliers_iqr(df_clean, col)
    print(f"\n{col}:")
    print(f"  Lower bound: {lower:.2f}")
    print(f"  Upper bound: {upper:.2f}")
    print(f"  Number of outliers: {len(outliers)} ({len(outliers)/len(df_clean)*100:.2f}%)")
    
    if len(outliers) > 0:
        print(f"  Outlier range: {outliers[col].min():.2f} to {outliers[col].max():.2f}")

# Create outlier visualization
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Outlier Detection - Box Plots', fontsize=16, fontweight='bold')

for i, col in enumerate(outlier_columns):
    axes[i].boxplot(df_clean[col].dropna(), patch_artist=True, 
                   boxprops=dict(facecolor='lightblue', alpha=0.7))
    axes[i].set_title(f'{col.replace("_clean", "").replace("_", " ").title()}')
    axes[i].set_ylabel('Value')
    axes[i].tick_params(axis='x', which='both', bottom=False, labelbottom=False)

plt.tight_layout()
plt.show()

# Option to cap outliers (commented out - can be enabled if needed)
# For extreme outliers, you can cap them:
# df_clean['discounted_price_clean'] = df_clean['discounted_price_clean'].clip(upper=df_clean['discounted_price_clean'].quantile(0.95))
print("\n💡 Note: Outliers detected but not removed. They may represent premium products or high-volume sellers.")
```

```python
# 9. PAIR PLOTS AND CORRELATION ANALYSIS
print("\n" + "="*60)
print("PAIR PLOTS AND CORRELATION ANALYSIS")
print("="*60)

# Select key numerical variables for pair plot
pair_plot_cols = ['discounted_price_clean', 'actual_price_clean', 'discount_percentage_clean', 'rating_clean', 'rating_count_clean']

# Create a subset for pair plot (sample if dataset is large)
if len(df_clean) > 1000:
    df_sample = df_clean[pair_plot_cols].sample(n=1000, random_state=42)
    print("Using sample of 1000 records for pair plot due to dataset size...")
else:
    df_sample = df_clean[pair_plot_cols]

# Create pair plot
print("Creating pair plot...")
plt.figure(figsize=(15, 12))
pair_plot = sns.pairplot(df_sample, diag_kind='hist', plot_kws={'alpha': 0.6})
pair_plot.fig.suptitle('Pair Plot - Key Numerical Variables', fontsize=16, y=1.02)
plt.show()

# Correlation heatmap
print("\nCorrelation Analysis:")
print("-" * 30)
correlation_matrix = df_clean[pair_plot_cols].corr()
print(correlation_matrix.round(3))

# Visualize correlation matrix
plt.figure(figsize=(10, 8))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm', center=0,
            square=True, fmt='.3f', cbar_kws={"shrink": .8})
plt.title('Correlation Heatmap - Key Variables', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Key correlations
print("\nKey Correlations:")
print("-" * 20)
significant_corr = correlation_matrix.abs() > 0.3
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        if significant_corr.iloc[i, j]:
            col1 = correlation_matrix.columns[i]
            col2 = correlation_matrix.columns[j]
            corr_val = correlation_matrix.iloc[i, j]
            print(f"{col1} ↔ {col2}: {corr_val:.3f}")
```

```python
# 11. CREATE CLEAN DATASET FOR TABLEAU
print("\n" + "="*60)
print("CREATING CLEAN DATASET FOR TABLEAU")
print("="*60)

# Create a new clean dataframe with essential columns for Tableau analysis
tableau_df = pd.DataFrame()

# Essential columns for dashboard
tableau_df['Product_ID'] = df_clean['product_id']
tableau_df['Product_Name'] = df_clean['product_name']

# Extract main category and subcategory
tableau_df['Main_Category'] = df_clean['category'].str.split('|').str[0]
tableau_df['Sub_Category'] = df_clean['category'].str.split('|').str[-1]
tableau_df['Full_Category'] = df_clean['category']

# Clean price columns
tableau_df['Discounted_Price'] = df_clean['discounted_price_clean']
tableau_df['Actual_Price'] = df_clean['actual_price_clean']
tableau_df['Discount_Percentage'] = df_clean['discount_percentage_clean']
tableau_df['Savings_Amount'] = df_clean['actual_price_clean'] - df_clean['discounted_price_clean']

# Rating information
tableau_df['Rating'] = df_clean['rating_clean']
tableau_df['Rating_Count'] = df_clean['rating_count_clean']

# Create derived metrics for analysis
tableau_df['Price_Range'] = pd.cut(tableau_df['Discounted_Price'], 
                                  bins=[0, 500, 1000, 2500, 5000, float('inf')], 
                                  labels=['Under ₹500', '₹500-1000', '₹1000-2500', '₹2500-5000', 'Above ₹5000'])

tableau_df['Rating_Category'] = pd.cut(tableau_df['Rating'], 
                                      bins=[0, 3.0, 3.5, 4.0, 4.5, 5.0], 
                                      labels=['Poor (≤3.0)', 'Fair (3.0-3.5)', 'Good (3.5-4.0)', 'Very Good (4.0-4.5)', 'Excellent (4.5-5.0)'])

tableau_df['Discount_Category'] = pd.cut(tableau_df['Discount_Percentage'], 
                                        bins=[0, 25, 50, 75, 100], 
                                        labels=['Low (0-25%)', 'Medium (25-50%)', 'High (50-75%)', 'Very High (75%+)'])

tableau_df['Review_Volume'] = pd.cut(tableau_df['Rating_Count'], 
                                    bins=[0, 100, 1000, 10000, float('inf')], 
                                    labels=['Low (<100)', 'Medium (100-1K)', 'High (1K-10K)', 'Very High (10K+)'])

# Add user and review information (limited)
tableau_df['User_ID'] = df_clean['user_id']
tableau_df['User_Name'] = df_clean['user_name']
tableau_df['Review_Title'] = df_clean['review_title']

# Clean up any remaining missing values
tableau_df = tableau_df.fillna({
    'Product_Name': 'Unknown Product',
    'Main_Category': 'Unknown',
    'Sub_Category': 'Unknown',
    'User_Name': 'Anonymous'
})

# Display dataset information
print(f"Clean Dataset Shape: {tableau_df.shape}")
print(f"Columns: {list(tableau_df.columns)}")

print(f"\nDataset Summary:")
print("-" * 30)
print(f"Total Products: {tableau_df['Product_ID'].nunique():,}")
print(f"Total Categories: {tableau_df['Main_Category'].nunique()}")
print(f"Total Sub-categories: {tableau_df['Sub_Category'].nunique()}")
print(f"Average Price: ₹{tableau_df['Discounted_Price'].mean():,.2f}")
print(f"Average Rating: {tableau_df['Rating'].mean():.2f}")
print(f"Total Users: {tableau_df['User_ID'].nunique():,}")

# Check for any remaining missing values
missing_check = tableau_df.isnull().sum()
if missing_check.sum() > 0:
    print(f"\nRemaining missing values:")
    print(missing_check[missing_check > 0])
else:
    print("\n✅ No missing values in clean dataset!")

# Display sample data
print(f"\nSample of Clean Dataset:")
print(tableau_df.head())

# Export to CSV for Tableau
output_path = "D:/Amazon Sales Analysis/CleanedAmazonSales.csv"
tableau_df.to_csv(output_path, index=False)
print(f"\n✅ Clean dataset exported to: {output_path}")
print(f"📊 Ready for Tableau import!")

print(f"\nFILE SUMMARY:")
print("=" * 20)
print(f"Original dataset: {df.shape[0]:,} rows × {df.shape[1]} columns")
print(f"Cleaned dataset: {tableau_df.shape[0]:,} rows × {tableau_df.shape[1]} columns")
print(f"Size reduction: {(1 - tableau_df.shape[1]/df.shape[1])*100:.1f}% fewer columns")
print(f"File saved as: CleanedAmazonSales.csv")
```
